우리 은행에서 달러 환율 증감 데이터 크롤링 하기

```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import Select
import time

# 1. 사이트 접속
driver = webdriver.Chrome()
driver.get('https://spot.wooribank.com/pot/Dream?withyou=FXXRT0014')
time.sleep(2)
```

```python
# 2.시작 = 2015년 11월 26일

full_data = []

for i in reversed(range(2015, 2025)):
    # (1) 2015년
    year = driver.find_element(By.ID, "START_DATEY")
    start_year = Select(year)
    start_year.select_by_value(str(i))

    # (2) 11월
    month = driver.find_element(By.ID, "START_DATEM")
    start_month = Select(month)
    start_month.select_by_value("11")

    # (3) 26일
    day = driver.find_element(By.ID, "START_DATED")
    start_day = Select(day)
    start_day.select_by_value("26")


    # 3. 끝 = 2025년 11월 25일

    # (1) 2025년
    year2 = driver.find_element(By.ID, "END_DATEY")
    end_year = Select(year2)
    end_year.select_by_value(str(i + 1))

    # (2) 11월
    month2 = driver.find_element(By.ID, "END_DATEM")
    end_month = Select(month2)
    end_month.select_by_value("11")

    # (3) 25일
    day2 = driver.find_element(By.ID, "END_DATED")
    end_day = Select(day2)
    end_day.select_by_value("25")

    # 4. 조회
    search_button = driver.find_element(By.CSS_SELECTOR, ".btn-area input").click()

    time.sleep(5)

    # 5. 날짜별 전체 정보

    daily_all = driver.find_elements(By.CSS_SELECTOR, ".tbl-type-1 tbody tr")


    # 6. 리스트 속 딕셔너리 형태 구성

    for all in daily_all:

        # (1) 날짜
        try:   
            date = all.find_element(By.CSS_SELECTOR, "td:nth-child(1)").text
        
        except:
            continue

        # (2) 환율 증감
        try:
            if all.find_element(By.CSS_SELECTOR, "td:nth-child(6) span").text == '상승' or '하락':
                
                raw_rate = all.find_element(By.CSS_SELECTOR, "td:nth-child(6)").text
                raw_rate2 = raw_rate.split('\n')[-1]
                rate = raw_rate2[2:]
            
            else:
                rate = '0'

        except:
            rate = "0"
        

        a = {"날짜" : date, "환율 증감" : rate}

        full_data.append(a)

full_data
```

```python
# 7. 전부 잘 크롤링 됐는지 확인

len(full_data)
```

```python
# 8. 엑셀화

import pandas as pd

df = pd.DataFrame(full_data)

df['날짜'] = pd.to_datetime(df['날짜'])
df['날짜'] = df['날짜'].dt.date

df.to_excel("우리 은행 환율 증감 데이터.xlsx", index=False)
```
